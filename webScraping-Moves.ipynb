```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

site="https://bulbapedia.bulbagarden.net/wiki/List_of_moves"
session = requests.Session()
headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:98.0) Gecko/20100101 Firefox/98.0",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
        "Accept-Language": "en-US,en;q=0.5",
        "Accept-Encoding": "gzip, deflate",
        "Connection": "keep-alive",
        "Upgrade-Insecure-Requests": "1",
        "Sec-Fetch-Dest": "document",
        "Sec-Fetch-Mode": "navigate",
        "Sec-Fetch-Site": "none",
        "Sec-Fetch-User": "?1",
        "Cache-Control": "max-age=0",
    }
response = session.get(site, headers=headers)
if response.status_code == 200:	
    web=response.text
    soup=BeautifulSoup(web)
    table=soup.table.tbody.tr.td.table.tbody
    print('Success!')
```

```python
rows=table.find_all("tr")
rows.pop(0)

name=[]
tipe=[]
cat=[]
pp=[]
pwr=[]
acc=[]
gen=[]

for row in rows:
    info=row.find_all("td")
    name.append(info[1].get_text(strip=True))
    tipe.append(info[2].get_text(strip=True))
    cat.append(info[3].get_text(strip=True))
    pp.append(info[4].get_text(strip=True))
    pwr.append(info[5].get_text(strip=True))
    acc.append(info[6].get_text(strip=True)[:-1])
    gen.append(info[7].get_text(strip=True))
```

```python
dict={
    "Name":name,
    "Type":tipe,
    "Category":cat,
    "PP":pp,
    "Power":pwr,
    "Accuracy":acc,
    "Gen":gen
}

df=pd.DataFrame.from_dict(dict)
```
